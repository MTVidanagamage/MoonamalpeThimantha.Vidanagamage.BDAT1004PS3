```python
#plotting line plot with labels
plt.plot(x, y_m,'r-', label='Marriages')
plt.plot(x, y_d,'b-', label='Divorces')
plt.legend()
plt.title("Marriages vs Divorces")
plt.xlabel("Year")
plt.ylabel("Marriages/Divorces")
plt.show()
```

```python

df_filtered = df.loc[(df['Year'] == 1900) | (df['Year'] == 1950) | (df['Year'] == 2000)]

x=df_filtered['Year'].values.astype('str')
y_m=df_filtered['Marriages_per_1000']
y_d=df_filtered['Divorces_per_1000']
```

```python
#plotting vertical bar chart with labels
plt.bar(x, y_m,label='Marriages')
plt.bar(x, y_d,label='Divorces')
plt.title("Marriages vs Divorces")
plt.xlabel("Year")
plt.ylabel("Marriages/Divorces")
plt.legend()
plt.xticks(x)
plt.show()
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['roman-emperor-reigns.csv']))
print(df)
```

```python
df = df.groupby(['Cause_of_Death'])['Cause_of_Death'].count().reset_index(name='Count')
print(df)
```

```python
import matplotlib.pyplot as plt
#plotting pie chart with labels
plt.pie(df['Count'], labels=df['Cause_of_Death'], autopct='%1.1f%%')
plt.title('Assassinated vs Not Assassinated')
plt.axis('equal')
plt.show()
```
